```python
import numpy as np


class HebbNetwork:
    def __init__(self, input_size, n=10, learning_rate=0.01):
        self.n = n
        self.weights = np.random.rand(input_size) * 0.01  
        self.learning_rate = learning_rate

    def normalize(self, inputs):
        return (inputs - np.min(inputs)) / (np.max(inputs) - np.min(inputs))

    def train(self, inputs):
        inputs = self.normalize(inputs)

        for i in range(self.n):
            for x in inputs:
                out = self.predict(x)
                self.weights += self.learning_rate * x * out

            self.weights /= np.linalg.norm(self.weights)
            print(f"epoch: {i}, weights: {self.weights}")

    def predict(self, x):
        return np.dot(self.weights, x)


network = HebbNetwork(input_size=3, n=1000)

training_data = np.random.rand(3, 3)

network.train(training_data)


test_input = np.random.rand(3)

output = network.predict(test_input)
print("    :", network.predict(test_input))
```
